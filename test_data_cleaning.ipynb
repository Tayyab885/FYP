```python
data_test.rbc = data_test.rbc.map({ 'normal' : 1, 'abnormal' : 0})
data_test.pc = data_test.pc.map({ 'normal' : 1, 'abnormal' : 0})
data_test.pcc = data_test.pcc.map({ 'present' : 1, 'notpresent' : 0})
data_test.ba = data_test.ba.map({ 'present' : 1, 'notpresent' : 0})
data_test.htn = data_test.htn.map({ 'yes' : 1, 'no' : 0})
data_test.dm = data_test.dm.map({ 'yes' : 1, '\tyes' : 1, '\tno' : 0, 'no' : 0})
data_test.cad = data_test.cad.map({ 'yes' : 1, 'no' : 0, '\tno' : 0})
data_test.appet = data_test.appet.map({ 'good' : 1, 'poor' : 0})
data_test.pe = data_test.pe.map({ 'yes' : 1, 'no' : 0})
data_test.ane = data_test.ane.map({ 'yes' : 1, 'no' : 0})
```

```python
data_test.pcv = pd.to_numeric(data_test.pcv, errors='coerce')
```

```python
def fill_nans_test():

    for col in data_test.columns:
        if data_test[col].dtype == 'float64':
            data_test[col] = data_test[col].transform(lambda x: x.fillna(x.median()))
    return data_test

fill_nans_test()
```
